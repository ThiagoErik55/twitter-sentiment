```python
# 2. Carregar dados da fonte original
try:
    df_raw = pd.read_csv('data/source/twitter_sentiment.csv', 
                         header=None, 
                         encoding='latin-1') 
    
    # Renomear as colunas
    df_raw.columns = ['Sentimento_Bruto', 'ID', 'Data_Bruta', 'FLAG', 'USER', 'Texto_Bruto']
    
    print(f"Dados carregados: {df_raw.shape[0]} linhas, {df_raw.shape[1]} colunas")

except Exception as e:
    print(f"ERRO DE LEITURA: {e}")
```

```python
# 4. Salvar na camada Bronze
df_raw.columns = ['Sentimento_Bruto', 'Data_Bruta', 'Texto_Bruto', 'FLAG', 'USER', 'TEXTO_ORIGINAL', 'data_ingestao', 'fonte_arquivo']

output_path = 'data/bronze/dados_brutos.csv' 
df_raw.to_csv(output_path, index=False, header=True)

print(f"\nDados brutos salvos com sucesso na camada Bronze: {output_path}")
```
