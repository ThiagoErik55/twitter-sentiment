Criação do schema

```python
# Script SQL para criar as tabelas
sql_script = """
-- 1. Tabela Principal de Fato (Armazena cada postagem limpa)
CREATE TABLE IF NOT EXISTS FATO_SENTIMENTO (
    id SERIAL PRIMARY KEY,
    -- Chave Estrangeira para a Data (Melhora a performance de query)
    data_postagem DATE NOT NULL,
    
    -- Métricas e Dados Principais
    score_sentimento DECIMAL(10,4),  -- O valor de -1.0 a 1.0 gerado na Silver Layer
    sentimento_binario INTEGER,       -- 0 ou 1 (Negativo/Positivo)
    texto_limpo TEXT,
    
    -- Metadados de Pipeline
    data_ingestao TIMESTAMP,
    data_processamento TIMESTAMP
);

-- 2. Tabela de Dimensão de Tempo (Crucial para o BI)
CREATE TABLE IF NOT EXISTS DIM_TEMPO (
    data DATE PRIMARY KEY,
    ano INTEGER,
    mes INTEGER,
    dia_da_semana VARCHAR(10)
);

-- 3. Tabela de Métricas Agregadas (Gold Layer)
CREATE TABLE IF NOT EXISTS METRICAS_DIARIAS (
    data_metrica DATE PRIMARY KEY,
    total_posts INTEGER,
    sentimento_medio DECIMAL(10,4),
    sentimento_max DECIMAL(10,4),
    sentimento_min DECIMAL(10,4)
);

-- 4. Criação de índices para performance (Opcional, mas profissional)
CREATE INDEX IF NOT EXISTS idx_fato_data ON FATO_SENTIMENTO (data_postagem);
"""
```
